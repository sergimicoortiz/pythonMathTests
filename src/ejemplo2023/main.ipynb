```python

import numpy as np
import scipy as sp
import sympy as sm
import fractions
np.set_printoptions(formatter={'all':lambda x: str(fractions.Fraction(x).limit_denominator())})
```

```python
a = np.array([
    [1,2,0],
    [-1,1,1],
    [1,0,1]
])
u=np.eye(3)
```

```python
# Apartado a
a_inv = np.linalg.inv(a)
a2 = np.matmul(a,a)
a3 = np.matmul(a2,a)
b= a3 - 3*a2 + 5*a
print(a_inv, '\n')
print(b, '\n')
```

```python
# Apartado b
b, c = sm.symbols('b c')
m = b*a2 + c*a + u
n = a_inv

def getEq(m,n):
    arr= []
    for i in range(len(m)):
        for j in range(len(m[i])):
            arr.append(sm.Eq(m[i][j], n[i][j]))
    return arr

res = sm.solve(getEq(m, n), [b,c])
print(res, '\n')
```
