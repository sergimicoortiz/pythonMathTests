```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from latex2sympy2 import latex2sympy
```

```python
x=np.linspace(10,0,20)
y=np.cos(x)*np.cos(x)
plt.plot(x,y)
```

```python
plt.plot(x,y, "o--")
```

```python
plt.figure(figsize=(20,5))
plt.plot(x,y, "o--", color="red", lw=1, ms=6)
plt.xlabel("test x")
plt.ylabel("test y")
```

```python
x2=np.linspace(10,0,20)
y2=np.sin(x)*np.sin(x)

plt.figure(figsize=(20,5))
plt.plot(x,y, "o--", color="red", lw=1, ms=6, label="l1")
plt.plot(x2,y2, "-", color="blue", lw=1, ms=6, label="l2")
plt.xlabel("test x")
plt.ylabel("test y")
plt.legend(loc="upper left")
```

$$
\begin{equation}
\int \frac{5 x^2-2 x+25}{x^3-6 x^2+25 x} d x
\end{equation}
$$

```python
x = sp.Symbol("x")
eq = latex2sympy(r"\int \frac{5 x^2-2 x+25}{x^3-6 x^2+25 x} d x")
f_x = sp.lambdify(x, eq.doit(), "numpy")
xx = np.linspace(0, 10, 100)
plt.plot(xx, f_x(xx))
```
